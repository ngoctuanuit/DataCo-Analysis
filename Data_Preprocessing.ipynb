```python
import pandas as pd
df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='latin1')
df.head()
```

```python
df['Customer Name'] = df['Customer Fname'] + ' ' + df['Customer Lname']
df['Customer Name'].head()
```

```python
df.columns
```

```python
columns = ['Order Id', 'Type', 'Order Status', 'order date (DateOrders)', 'Sales', 'Market', 'Order Item Discount Rate',
           'Latitude', 'Longitude', 'Order Item Quantity',
           'Category Id', 'Category Name', 
           'Days for shipping (real)', 'Days for shipment (scheduled)', 'Delivery Status', 'Shipping Mode',
           'Customer Id', 'Customer Country', 'Customer Name', 'Customer Segment',
           'Product Card Id', 'Product Name', 'Product Price'
           ]
# Table: Orders, Customers, Categories, Prouducts, Shipping, Time
```

```python
df_res = df[columns]
df_res.head()
```

```python
df_res.info()
```

```python
df_res.dropna(inplace=True)
df_res = df_res.drop_duplicates()
df_res.info()
```

```python
df_res.head()
```

```python
df_res.columns
```

```python
df_res = df_res.rename(columns={'Order Item Discount Rate':'discount_rate', 'Order Item Quantity': 'product_quantity',
                                'order date (DateOrders)': 'order_date', 'Days for shipping (real)': 'day_shipping_real',
                                'Days for shipment (scheduled)':'day_shipment_scheduled', 'Product Card Id': 'product_id'})
```

```python
df_res.columns
```

```python
df_res.columns = df_res.columns.str.lower().str.replace(' ', '_')
df_res.head()
```

```python
df_res.columns
```

```python
df_res.info()
```

```python
df_res['shipping_id'] = range(1, len(df_res) + 1)
```

```python
df_res.columns
```

```python
df_product = df[['product_id', 'product_name', 'product_price', 'category_id']]
df_product = df_product.drop_duplicates(subset=['product_id'])
df_product.to_excel('dim_product.xlsx', index = False)
```

```python
df_category = df[['category_id', 'category_name']]
df_category = df_category.drop_duplicates(subset=['category_id'])
df_category.to_excel('dim_category.xlsx', index = False)
```

```python
df_customer = df[['customer_id', 'customer_country', 'customer_name', 'customer_segment']]
df_customer = df_customer.drop_duplicates(subset=['customer_id'])
df_customer.to_excel('dim_customer.xlsx', index = False)
```

```python
df_shipping = df[['shipping_id', 'shipping_mode','day_shipping_real', 'day_shipment_scheduled', 'delivery_status']]
df_shipping.to_excel('dim_shipping.xlsx', index = False)
```

```python
df.order_date = pd.to_datetime(df.order_date)
df_date = df[['order_date']]
df_date['order_month'] = df.order_date.dt.month
df_date['order_quarter'] = df.order_date.dt.quarter
df_date['order_year'] = df.order_date.dt.year
df_date.to_excel('dim_date.xlsx', index = False)
```

```python
df_fact = df[['order_id', 'product_id', 'customer_id', 'shipping_id', 'order_date', 'type', 
              'order_status', 'sales', 'market', 'discount_rate', 'product_quantity']]
df_fact.to_excel('fact.xlsx', index = False)
```
